```python
import gc
import numpy as np
from PIL import Image
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Function to load and preprocess data
def load_data_in_batches(data_dir, categories, target_size=(224, 224), batch_size=32):
    # Use ImageDataGenerator to load and process images in batches
    data_gen = ImageDataGenerator(rescale=1.0/255.0)  # Normalize images
    # Load images using the flow_from_directory method
    generator = data_gen.flow_from_directory(
        data_dir,
        target_size=target_size,  # Resize images
        batch_size=batch_size,
        class_mode='sparse',  # Labels are integers (sparse categorical)
        classes=categories  # Define the classes to load
    )
    return generator

# Define dataset paths
train_dir = r"C:\Users\yashu\Desktop\Dataset\train"
test_dir = r"C:\Users\yashu\Desktop\Dataset\test"

# Define categories
categories = ['organic', 'recycle', 'unknown']

# Load data in batches for training
train_generator = load_data_in_batches(train_dir, categories)

# Load data in batches for testing (if required)
test_generator = load_data_in_batches(test_dir, categories)

# Print the information about the training data
print(f"Training data info: {train_generator.class_indices}")
print(f"Training data batches: {train_generator.samples} images in {train_generator.batch_size} batch size")
print(f"Test data info: {test_generator.class_indices}")
print(f"Test data batches: {test_generator.samples} images in {test_generator.batch_size} batch size")

# Manually run garbage collection after loading data
gc.collect()  # Clean up memory

# Print memory freed
print(f"Memory freed: {gc.collect()}")  # gc.collect() returns the number of unreachable objects

# Now you can continue with your model training
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create ImageDataGenerator for training data with augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0/255.0,           # Normalize images
    rotation_range=40,           # Random rotations
    width_shift_range=0.2,       # Random horizontal shifts
    height_shift_range=0.2,      # Random vertical shifts
    shear_range=0.2,             # Random shearing
    zoom_range=0.2,              # Random zoom
    horizontal_flip=True,        # Random horizontal flips
    fill_mode='nearest'          # Fill pixels after transformation
)

# Specify the correct path to your training data
train_generator = train_datagen.flow_from_directory(
    r'C:\Users\yashu\Desktop\Dataset\train',  # Path to your training data
    target_size=(224, 224),                   # Resize images to 224x224
    batch_size=32,                            # Batch size
    class_mode='categorical'                  # For multi-class classification
)

# Example to check if it works:
print(train_generator.class_indices)  # Print class labels
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Define the CNN model
model = Sequential()

# Add a convolutional layer with 32 filters, kernel size 3x3, activation 'relu'
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(MaxPooling2D(2, 2))  # Max pooling with pool size of 2x2

# Add another convolutional layer
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))

# Add another convolutional layer
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))

# Flatten the output to connect to the dense layers
model.add(Flatten())

# Add a fully connected layer (Dense layer)
model.add(Dense(512, activation='relu'))

# Add dropout for regularization
model.add(Dropout(0.5))

# Add the output layer with 3 classes (for 'organic', 'recycle', 'unknown')
model.add(Dense(3, activation='softmax'))

# Summarize the model architecture
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',  # Optimizer
              loss='categorical_crossentropy',  # Loss function for multi-class classification
              metrics=['accuracy'])  # Evaluation metric
```

```python
# Train the model using the train_generator 
history = model.fit(
    train_generator,  # Data generator
    steps_per_epoch=train_generator.samples // train_generator.batch_size,  # Number of steps per epoch
    epochs=15,  # Number of epochs to train the model
    verbose=1
)
```

```python
# Define the test data generator
test_datagen = ImageDataGenerator(rescale=1.0/255.0)  # Rescale images for test data

test_generator = test_datagen.flow_from_directory(
    r'C:\Users\yashu\Desktop\Dataset\test',  # Path to test data
    target_size=(224, 224),  # Resize images to 224x224
    batch_size=32,           # Batch size
    class_mode='categorical'  # For multi-class classification
)

# Evaluate the model on the test data
test_loss, test_acc = model.evaluate(test_generator, steps=test_generator.samples // test_generator.batch_size, verbose=1)
print(f"Test Accuracy: {test_acc * 100:.2f}%")
 
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# Path to the image
image_path = r"C:\Users\yashu\Pictures\Saved Pictures\example_image.jpg"

# Load the image
img = image.load_img(image_path, target_size=(224, 224))  # Resize to model's input size
img_array = image.img_to_array(img)  # Convert the image to a numpy array
img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
img_array = img_array / 255.0  # Normalize the image (same as model training)

# Make prediction
predictions = model.predict(img_array)

# Get the predicted class index (highest probability)
predicted_class_index = np.argmax(predictions, axis=-1)[0]

# Class labels (make sure they match your model's output classes)
class_labels = ['organic', 'recycle', 'unknown']

# Get the predicted class and confidence
predicted_class = class_labels[predicted_class_index]
confidence = np.max(predictions) * 100  # Convert to percentage

# Display the prediction and confidence
print(f"Predicted Class: {predicted_class}")
print(f"Prediction Confidence: {confidence:.2f}%")

# Optionally, display the image
plt.imshow(img)
plt.title(f"Predicted: {predicted_class} ({confidence:.2f}%)")
plt.axis('off')
plt.show()
```
